```python
import pandas as pd

c.execute("SELECT DISTINCT * FROM halo_db_2")
z = c.fetchall()

halodf = pd.DataFrame(z,columns = ["Name","Categories","length of page"])
```

```python
# 10 Longest pages on the wiki
halodf.sort_values(by = "length of page",ascending=False).head(10)
```

```python
# 10 Shortest pages on the wiki
halodf.sort_values(by = "length of page",ascending=False).tail(10)
```

```python
halovehicles = halodf[halodf["Categories"].str.contains("Vehicles")]
halovehicles.sort_values(by = "length of page",ascending=False).head(10)
```
